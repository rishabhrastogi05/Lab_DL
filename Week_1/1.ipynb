```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mnist = tf.keras.datasets.fashion_mnist
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

## Number of Classes

```python
print(f"Number of classes : {len(np.unique(Y_train))}")
```

## Dimensions of each image

```python
print(X_train[0].shape)
```

## One Image from each class

```python
for i in range(0,10):
  idx = Y_train.tolist().index(i)
  plt.imshow(X_train[idx])
  plt.show()
```

```python
print(np.max(X_train[0]), np.min(X_train[0]), sep="\n")
```

## Normalizing the Data

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

## Building a 2 Layer Neural Network

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=[28,28], name="inputLayer"))
model.add(tf.keras.layers.Dense(350, activation= 'relu', name ='HiddenLayer1'))
model.add(tf.keras.layers.Dense(165, activation= 'relu', name ='HiddenLayer2'))
model.add(tf.keras.layers.Dense(10, activation= 'softmax', name ='outputLayer'))
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, batch_size = 16, epochs=20, validation_split=0.1)
```

```python
acc = {}
acc['model_1'] = model.evaluate(X_test,Y_test)
```

## Building a 5-Layer Neural Network

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=[28,28], name="inputLayer"))
model.add(tf.keras.layers.Dense(450, activation= 'relu', name ='HiddenLayer1'))
model.add(tf.keras.layers.Dense(225, activation= 'relu', name ='HiddenLayer2'))
model.add(tf.keras.layers.Dense(95, activation= 'relu', name ='HiddenLayer3'))
model.add(tf.keras.layers.Dense(45, activation= 'relu', name ='HiddenLayer4'))
model.add(tf.keras.layers.Dense(22, activation= 'relu', name ='HiddenLayer5'))
model.add(tf.keras.layers.Dense(10, activation= 'softmax', name ='outputLayer'))
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, batch_size = 16, epochs=20, validation_split=0.1)
```

```python
acc['model_2'] = model.evaluate(X_test, Y_test)
```

```python
print("Accuracy : ",f"Model 1 : {acc['model_1'][1]}", f"Model 2 : {acc['model_2'][1]}", sep="\n")
```

## We can see model 1 which has 2 hidden layers provides better accuracy than model 2 which has 5 hidden layers.
